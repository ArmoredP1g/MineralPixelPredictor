```python
def interpolate(x_list, y_list, x):
    # 根据x插值光谱数据
    result = []
    idx = 0
    for i in range(x.__len__()):
        while not (x_list[idx] <= x[i] < x_list[idx+1]):
            idx += 1
        result.append(y_list[idx+1]-(x_list[idx+1]-x[i])*((y_list[idx+1]-y_list[idx])/(x_list[idx+1]-y_list[idx])))
    return result
    
```
